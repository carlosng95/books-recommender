```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep = ';',on_bad_lines = 'skip', encoding = 'latin1')
print(ratings.shape)
ratings.head()
```

```python
ratings.rename(columns = {
    'User-ID': 'user_id',
    'Book-Rating': 'rating'
}, inplace = True)
```

```python
ratings.user_id.value_counts()
```

```python
# Only users who had at least rated more than 200 books
x = ratings.user_id.value_counts() > 200
x
```

```python
x[x].shape
```

```python
y = x[x].index.tolist()
y
```

```python
ratings = ratings[ratings.user_id.isin(y)]
```

```python
ratings.head()
```

```python
# join ratings with book

ratings_with_books = ratings.merge(books, on = 'ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.groupby('author')['rating'].agg(['mean', 'count']).sort_values(by='count', ascending=False)
```

```python
ratings_with_books.groupby('year')['rating'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
```

```python
number_rating = ratings_with_books.groupby('title').rating.count().reset_index()
number_rating
```

```python
number_rating.columns = ['title','num_of_rating']
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on = 'title')
```

```python
final_rating.head()
```

```python
# lets take those books which got al least 50 rating of users
final_rating = final_rating[final_rating.num_of_rating >= 50]

final_rating.drop_duplicates(['user_id','title'], inplace = True)
```

```python
final_rating.shape
```

```python
final_rating.isnull().sum()
```

```python
final_rating.head()
```

```python
final_rating.nunique()
```

```python
book_pivot = final_rating.pivot_table(
    index='title',
    columns='user_id',
    values='rating'
)
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```
