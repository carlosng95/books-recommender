```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('BX-Books.csv', sep = ';',on_bad_lines = 'skip' ,encoding = 'latin1', low_memory=False)
```

```python
books = books.iloc[:,[0,1,2,3,4,-1]]
books.head()
```

```python
books.rename(columns = {
    'Book-Title': 'title',
    'Book-Author': 'author',
    'Year-Of-Publication': 'year',
    'Publisher': 'publisher',
    'Image-URL-L': 'image_url'
}, inplace = True)
```

```python
users = pd.read_csv('BX-Users.csv', sep = ';',on_bad_lines = 'skip', encoding = 'latin1')
print(users.shape)
users.head()
```

```python
users.rename(columns = {
    'User-ID': 'user_id',
    'Location': 'location',
    'Age': 'age'
}, inplace = True)
```

```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep = ';',on_bad_lines = 'skip', encoding = 'latin1')
print(ratings.shape)
ratings.head()
```

```python
ratings.rename(columns = {
    'User-ID': 'user_id',
    'Book-Rating': 'rating'
}, inplace = True)
```

```python
ratings.user_id.value_counts()
```

```python
# Only users who had at least rated more than 200 books
x = ratings.user_id.value_counts() > 200
x
```

```python
x[x].shape
```

```python
y = x[x].index.tolist()
y
```

```python
ratings = ratings[ratings.user_id.isin(y)]
```

```python
ratings.head()
```

```python
# join ratings with book

ratings_with_books = ratings.merge(books, on = 'ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.groupby('author')['rating'].agg(['mean', 'count']).sort_values(by='count', ascending=False)
```

```python
ratings_with_books.groupby('year')['rating'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
```

```python
number_rating = ratings_with_books.groupby('title').rating.count().reset_index()
number_rating
```

```python
number_rating.columns = ['title','num_of_rating']
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on = 'title')
```

```python
final_rating.head()
```

```python
# lets take those books which got al least 50 rating of users
final_rating = final_rating[final_rating.num_of_rating >= 50]

final_rating.drop_duplicates(['user_id','title'], inplace = True)
```

```python
final_rating.shape
```

```python
final_rating.isnull().sum()
```

```python
final_rating.head()
```

```python
final_rating.nunique()
```

```python
book_pivot = final_rating.pivot_table(
    index='title',
    columns='user_id',
    values='rating'
)
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors = 6)
```

## Testing model

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors = 6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f'You searched {book_name}\n')
                print('The suggestion books are: \n')
            else:
                print(j)
```

```python
recommend_book('2nd Chance')
```
